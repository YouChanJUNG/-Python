```python
#1번 방법
i = 0
result1 = 0
while i < 100:
  i += 1
  if i % 2 == 0:
    result1 += i

print('1번 방법 : {0}'.format(result1))
```

```python
#2번 방법
j = 0
result2 = 0
while True:
  if j > 100:
    break

  j +=1
  if j % 2 == 0:
    result2 += j

print('2번 방법 (break) : {0}'.format(result2))
```

```python
#3번 방법
k = 0
result3 = 0
while k < 100:
  k += 1
  if k % 2 != 0:
    continue

  result3 += k

print('3번 방법 (continue) : {0}'.format(result3))
```
